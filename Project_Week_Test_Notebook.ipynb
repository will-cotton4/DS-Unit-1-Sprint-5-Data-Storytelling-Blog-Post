```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

updated_wdi_url = 'https://raw.githubusercontent.com/will-cotton4/DS-Unit-1-Sprint-5-Data-Storytelling-Blog-Post/master/Updated%20WDI%20Data.csv'

wdi = pd.read_csv(updated_wdi_url).drop(columns = ['Series Code', 'Country Code'])

short_years = {'2009 [YR2009]': 2009, '2010 [YR2010]': 2010, '2011 [YR2011]': 2011, '2012 [YR2012]': 2012,
       '2013 [YR2013]': 2013, '2014 [YR2014]': 2014, '2015 [YR2015]': 2015, '2016 [YR2016]': 2016,
       '2017 [YR2017]': 2017, '2018 [YR2018]': 2018, '1999 [YR1999]': 1999, '2000 [YR2000]': 2000,
       '2001 [YR2001]': 2001, '2002 [YR2002]': 2002, '2003 [YR2003]': 2003, '2008 [YR2008]': 2008,
       '2007 [YR2007]': 2007, '2006 [YR2006]': 2006, '2005 [YR2005]': 2005, '2004 [YR2004]': 2004}
wdi = wdi.rename(columns = short_years)

short_names = {"GNI per capita, PPP (current international $)":"GNI per capita",
              "Human capital index (HCI) (scale 0-1)": "HCI",
              "Foreign direct investment, net inflows (BoP, current US$)": "FDI",
              "Agriculture, forestry, and fishing, value added (% of GDP)": "Agriculture pct of GDP",
              "Poverty headcount ratio at $1.90 a day (2011 PPP) (% of population)": "Poverty (1.90)",
              "Domestic credit provided by financial sector (% of GDP)": "Domestic credit pct of GDP",
              "Poverty headcount ratio at national poverty lines (% of population)": "Poverty (natl lines)"}
wdi = wdi.replace(short_names)
wdi = wdi.replace("..",np.NaN)

# All of the data for 2018 is missing:
wdi = wdi.drop(columns=[2018])
for column in range(1999, 2018):
  wdi[column] = pd.to_numeric(wdi[column])
```
